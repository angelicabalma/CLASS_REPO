```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
city_ride = ride_df.merge(city_df, how="left", on="city")
city_ride.head()
# Display the data table for preview
```

```python
#Average Fare ($) Per City
#Total Number of Rides Per City
#Total Number of Drivers Per City
#City Type (Urban, Suburban, Rural)
groupby = city_ride.groupby(["city"])
avg_fare = groupby["fare"].mean()
total_rides = groupby["ride_id"].count()
total_drivers = groupby["driver_count"].sum()


urban_cities = city_ride[city_ride['type']== 'Urban']
groupby = urban_cities.groupby(["city"])
avg_urban = groupby["fare"].mean()
total_urban = groupby["ride_id"].count()
driver_urban = groupby["driver_count"].sum()



rural_cities = city_ride[city_ride['type']== 'Rural']
groupby = rural_cities.groupby(["city"])
avg_rural = groupby["fare"].mean()
total_rural = groupby["ride_id"].count()
driver_rural = groupby["driver_count"].sum()


suburban_cities = city_ride[city_ride['type']== 'Suburban']
groupby = suburban_cities.groupby(["city"])
avg_suburban = groupby["fare"].mean()
total_suburban = groupby["ride_id"].count()
driver_suburban = groupby["driver_count"].sum()

plt.scatter(total_urban, avg_urban, color="Coral", marker="o", edgecolors="black", s=driver_urban, alpha=0.75)
plt.scatter(total_rural, avg_rural, color="Gold", marker="o", edgecolors="black", s=driver_rural, alpha=0.75)
plt.scatter(total_suburban, avg_suburban, color="SkyBlue", marker="o", edgecolors="black", s=driver_suburban, alpha=0.75)

label_colours = ('Urban', 'Rural', 'Suburban')
colours = ('Coral', 'Gold', 'SkyBlue')
legend = plt.legend(labels=label_colours)
plt.xlim = (0,40)
plt.ylim = (30, 50)


plt.title('Pyber Ride Sharing Data 2018')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.grid(True)
legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]
plt.text(45, 41, 'Note: Circle size correlates with driver count per city')
plt.savefig("../Pyber/rida_sharing_scatter.png", bbox_inches='tight')
plt.show()

```

## Total Fares by City Type

```python
# Calculate Type Percents
groupby = city_ride.groupby(["type"]).sum()['fare']
total_count = city_ride['fare'].sum()
percentage = (groupby*100)/total_count
labels = ['Rural', 'Suburban', 'Urban']
colours = ['Gold', 'SkyBlue', 'Coral']
explode =[0, 0, 0.1]
plt.pie(percentage, explode=explode, colors=colours,labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=150)
plt.title('% Of Total Fares by City Type')
plt.savefig("../Pyber/total_fares.png")
plt.show()
```
